# COVID-19 Data Exploratory Analysis

```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing dataset
df=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df.tail()
```

```python
#dropping the last row which has the total counts and the column sno
df=df.iloc[:-1,1:]
df.tail()
```

```python
#plotting statewise positive cases
plt.figure(figsize=(20,10))
plt.barh(df['state_name'],df['positive'],color='orange')
plt.ylabel('States')
plt.xlabel('Positive Cases')
plt.title('Statewise Total Positive Cases')
```

```python
#plotting statewise active cases
plt.figure(figsize=(20,10))
plt.barh(df['state_name'],df['active'],color='yellow')
plt.ylabel('States')
plt.xlabel('Active Cases')
plt.title('Statewise Active Caseload')
```

```python
#plotting statewise cured cases
plt.figure(figsize=(20,10))
plt.barh(df['state_name'],df['cured'],color='green')
plt.ylabel('States')
plt.xlabel('Cured Cases')
plt.title('Statewise Cured Cases')
```

```python
#plotting statewise deathload
plt.figure(figsize=(20,10))
plt.barh(df['state_name'],df['death'],color='red')
plt.ylabel('States')
plt.xlabel('Total Number of Deaths')
plt.title('Statewise Total deaths')
```
